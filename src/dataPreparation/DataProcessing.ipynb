```python
import pandas as pd
import re
import os
import nltk
os.chdir("/home/rethik/QuestionTagging/")
```

```python
rawdata = pd.read_csv("Data/rawdata.csv")
```

```python
rawdata.head()
```

```python
rawdata["text"] = rawdata["title"] + " " + rawdata["body"]
rawdata.drop(["title","body"],axis= 1, inplace = True)
```

```python
rawdata.head()
```

```python
def clean_text(text):
    text = re.sub(r'<.*?>', "", str(text))
    text = re.sub(r'a-zA-z', " ", text)
    text = " ".join(text.split())
    text = text.lower()
    
    words = [word for word in text.split() if word not in stop_words]
    clean_text = " ".join(words)
    
    return clean_text
```

```python
stop_words = set(nltk.corpus.stopwords.words('english'))
rawdata["cleaned_text"] = rawdata["text"].apply(lambda x:clean_text(x))
```
